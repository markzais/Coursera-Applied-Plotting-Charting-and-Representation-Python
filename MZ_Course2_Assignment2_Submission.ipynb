```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    #df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))
    df = pd.read_csv('BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'b091d88a3b4e7d3c9707df0e1cd8e7d5be82221e8f40485f1464c1ae')
```

```python
def format_data():
    
    """ reads in the data file, creates DayOfYear, Year and Temperature columns
    """
    
    # read the data
    #df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/b091d88a3b4e7d3c9707df0e1cd8e7d5be82221e8f40485f1464c1ae.csv')
    df = pd.read_csv('b091d88a3b4e7d3c9707df0e1cd8e7d5be82221e8f40485f1464c1ae.csv')
    
    # Convert Date column to date data type
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
    
    # Create MonthDay and Year columns
    df['MonthDay'] = df['Date'].dt.strftime('%m/%d')
    df['Year'] = df['Date'].dt.year
    
    # Create a separate column for the month
    df['Month'] = pd.DatetimeIndex(df['Date']).month
    
    # Remove February 29th dates
    df = df[df.MonthDay != '02/29']
    
    # Sort by the Date column
    df = df.sort_values('Date')
    
    # Convert temperature values
    df['Temperature'] = df['Data_Value']/10.0
    
    # Delete the Data_Value and Date columns
    del df['Data_Value']
    del df['Date']
    
    return df
```

```python
import numpy as np   

df = format_data()
# Make separate dataset for 2015
df_15 = df[df.Year == 2015]

# Remove data from 2015 and later
df = df[df.Year < 2015]

# Get the max and min for each day
max_temp = df[df['Element'] == 'TMAX'].groupby('Month').aggregate({'Temperature':np.max})
min_temp = df[df['Element'] == 'TMIN'].groupby('Month').aggregate({'Temperature':np.min})

#max_temp_15 = df_15[df_15['Element'] == 'TMAX'].groupby('MonthDay').aggregate({'Temperature':np.max})
#min_temp_15 = df_15[df_15['Element'] == 'TMIN'].groupby('MonthDay').aggregate({'Temperature':np.min})

record_max = df_15[df_15['Temperature'] > df_15['Month'].map(max_temp['Temperature'])]
record_min = df_15[df_15['Temperature'] < df_15['Month'].map(min_temp['Temperature'])]

plt.figure()

plt.plot(max_temp, '-o', label='Max Temperature (2005-2014)', color='tomato')
plt.plot(min_temp, '-o', label='Min Temperature (2005-2014)', color='teal')

plt.fill_between(max_temp.index.values,min_temp.values.flatten(),max_temp.values.flatten(), facecolor='grey',alpha='0.5')

plt.scatter(record_max.Month.tolist(),record_max['Temperature'],marker='*',s=20*10,c='darkred',label='Record High (2015)') 
plt.scatter(record_min.Month.tolist(),record_min['Temperature'],marker='*',s=20*10,c='darkblue',label='Record Low (2015)') 

# Range is 1 to 12 with intervals of 1
plt.xticks(range(1, len(min_temp)+1, 1))
#plt.xticks(list(max_temp.index),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec']) 

plt.legend(frameon = False)

plt.xlabel('Month',fontsize=20)
plt.ylabel('Temperature ($^\circ$C)',fontsize=20)
plt.title('Assignment 2. Temperatures for Valrico, Florida, United States',fontsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
fig = plt.gcf()
fig.set_size_inches(20,10)

plt.show()
```
